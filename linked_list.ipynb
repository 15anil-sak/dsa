```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Reverse the linked list using a stack
def reverse(head):
    if not head or not head.next:
        return head

    stack = []
    curr = head
    while curr:
        stack.append(curr)
        curr = curr.next

    head = stack.pop()  # New head (last node)
    curr = head
    while stack:
        curr.next = stack.pop()
        curr = curr.next
    curr.next = None  # Terminate the list

    return head

# Print the linked list
def PrintList(head):
    curr = head
    while curr:
        print(curr.data, '->', end=' ')
        curr = curr.next
    print("None")

# Test the code
n1 = Node(1)
n1.next = Node(2)
n1.next.next = Node(3)

print("Original List:")
PrintList(n1)

# Reverse the list
n1 = reverse(n1)

print("Reversed List:")
PrintList(n1)
```
